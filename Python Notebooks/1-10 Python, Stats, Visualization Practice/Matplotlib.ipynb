```python
import matplotlib.pyplot as plt

%matplotlib inline

x = list(range(0, 10))
y = list(range(-10, 0))

plt.plot(x, y)
```

```python
a = [0, -100, 25, 67, -323]
b = [0, 3, 7, 3, 9]

plt.plot(a, b)
```

```python
plt.axis([-50, 80, 2, 8])

plt.plot(a, b)
```

```python
plt.title("Triangle")
plt.xlabel("Array A")
plt.ylabel("Array B")
plt.axis([-50, 80, 2, 8])
plt.xticks((-40, -20, 0, 20, 40, 60, 80), ('h', 'e', 'l', 'l', 'o', 'o', '0'))

plt.plot(a, b, color="red")
```
