```python
import pandas as pd
```

```python
ufo = pd.read_csv('http://bit.ly/uforeports')
```

### Pandas DataFrames loc

```python
ufo.head(3)
```

```python
ufo.loc[ 0, :]
```

```python
ufo.loc[[0,1,2], :]
```

```python
ufo.loc[0:2, :]   # Same
```

```python
ufo.loc[:, ['City', 'State']]
```

```python
ufo.loc[:, 'City':'State']
```

```python
ufo.loc[0:2, 'City':'State']
```

```python
ufo.head(3).drop('Time', axis=1)   # Same
```

```python
ufo[ufo.City=='Oakland']
```

```python
ufo.loc[ufo.City=='Oakland', :]   # Same
```

```python
ufo.loc[ufo.City=='Oakland', 'State']
```
