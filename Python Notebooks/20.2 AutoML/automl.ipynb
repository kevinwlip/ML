Automated Machine Learning (AutoML) automating the discovery of hte best machine learning algorithms and hyperparameters in order to solve predictive modeling tasks with very little to no user involvement.

Auto-sklearn is an open-source library for performing AutoML in Python and it uses the Scikit-Learn Python package for transformations and machine learning algorithms as well as a Bayesian Optimization search procedure. This is used to discover the top-performing modle pipeline for a given dataset.

```python
# Debugging and Formatter

from IPython.core.debugger import set_trace

% load_ext nb_black

import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_classification
import random

random.seed(0)
min_class = random.uniform(0.1, 0.3)

X, y = make_classification(
    n_samples=100000,
    n_features=5,
    n_redundant=0
    n_informative=5,
    n_repeated=0,
    n_classes=2,
    weights=[min_class, 1 - min_class],
    flip_y=0,
    random_state=7
)
```

```python
from collections import Counter

counter = Counter(y)

plt.figure(1, figsize=(16,4))
plt.bar(counter.keys(), counter.values())
```

**Train Test Split**

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

**Baseline Logistic Regression**

Let's first run a baseline Logistic Regression model to have a benchmark

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
model = LogisticRegression()

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

acc = accuracy_score(y_test, y_pred)
print(f"Accuracy: {acc}")
```

**AutoML**

Since our task is classification we will create and configure an instance of the AutoSklearn=Classifier class and fit it on the dataset, just as we did before

```python
from autosklearn.classification import AutoSklearnClassifier
```

By default, the search for the best performing model will use the train test split and will run for an hour.

We can tweak that by setting the time left for this task argument to 3 minutes for this example.

Also we'll set the time limit for fitting a machine learning model high enough, we'll set it to 30 seconds although this problem should definitely take less time.

Also by default, the search creates an ensemble of top-performing models discovered as part of the search. Many times this leads to overfitting so I usually disable this. But let's try with and without.

```python
model = AutoSklearnClassifier(
    time_left_for_this_task=3 * 60,
    per_run_time_limit=30,
    ensemble_size=1,
    initial_configurations_via_metalearning=0,
    n_jobs=-1
)
```

```python
%%time
model.fit(X_train, y_train)
```

```python
print(model.sprint_statistics())
```

```python
print(model.show_models())
```

```python
y_pred = model.predict(X_test)
acc = accuracy_score(y_test, y_pred)
print(f"Accuracy: {acc}")
```

For regression we have a similar approach although we have to use AutoSklearnRegressor instead of AutoSklearnClassifier.

We can create a regression problem with the make_regression class instead of make_classification from scikit-learn.

Also the metric will have to be regression specific (ex. mean_absolute_error).
